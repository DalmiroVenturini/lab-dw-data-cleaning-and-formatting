## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*


```python
###EXERCISE 1: CLEANING COLUMN NAMES

import pandas as pd

# Load Titanic dataset from an online source
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
company_df = pd.read_csv(url)
company_df
```

```python
#Cambiar formato en columnas: todo minúscula, reemplazar " " por "_"

company_df.columns = [column.lower().replace(" ", "_") for column in company_df.columns]
company_df
```

```python
#Reemplazar "st" por "state"

company_df1 = company_df.rename(columns = {"st":"state"})
company_df1
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
### EXERCISE 2: CLEANING INVALID VALUES

#Ver valores posibles en Gender

company_df1["gender"].unique()
```

```python
#Estandarización de atributo "Gender"

company_df1["gender"] = company_df1["gender"].replace({"Male":"M", "Femal":"F", "female":"F"})
company_df1
```

```python
#Ver valores posibles en state

company_df1["state"].unique()
```

```python
#Reemplazo las abreviaciones en los nombres de los estados por sus nombres completos

company_df1["state"] = company_df1["state"].replace({"WA":"Washington", "AZ":"Arizona", "Cali":"California"})
company_df1
```

```python
#En educación reemplazo Bachelors por Bachelor

company_df1["education"] = company_df1["education"].replace({"Bachelors":"Bachelor"})
company_df1
```

```python
#En customer_lifetime_value, elimino el %.
#Convierto los valores a numéricos con astype también

company_df1["customer_lifetime_value"] = company_df1["customer_lifetime_value"].str.replace("%","").astype(float)
company_df1
```

```python
#En educación reemplazo Bachelors por Bachelor

company_df1["vehicle_class"] = company_df1["vehicle_class"].replace({"Sports Car":"Luxury", "Luxury SUV":"Luxury", "Luxury Car":"Luxury"})
company_df1
```

```python
#Comprobamos valores posibles en columna vehicle_class

company_df1["vehicle_class"].unique()
```

```python
### EXERCISE 4: DEALING WITH NULL VALUES

#Chequeo cuantos null hay en cada columna

company_df1.isnull().sum()
```

```python
#Hay muchos valores nulos. Entonces elimino las filas que tengan TODOS los valores nulos

company_df2 = company_df1.dropna(how="all")
company_df2
```

```python
#Chequeo cuantos null hay ahora en cada columna

company_df2.isnull().sum() #Output: Vemos que en gender tenemos 117 valores nulos
```

```python
#Chequeo cuantos null hay en cada categoría de gender

company_df2["gender"].value_counts() #Output: Vemos que son parecidas las cantidades, entonces aplicamos un fill basado en frecuencias.
```

```python
#Reemplazo los valores nulos por valores random, ya que son muy parecidas las proporciones de ocurrencia entre M y F
import numpy as np #para usar la función random
company_df3 = company_df2.copy()
company_df3["gender"] = company_df3["gender"].fillna(pd.Series(np.random.choice(["M","F"], size=len(company_df3))))
company_df3.isnull().sum()
```

```python
### EXERCISE 5: DEALING WITH DUPLICATES

#Eliminar los 3 duplicados que quedan en la columna "customer_lifetime_value"

company_df3 = company_df3.dropna(subset=["customer_lifetime_value"])
company_df3.isnull().sum()
```
